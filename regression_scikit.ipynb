```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
# vecteur colonne de feature
X = 2 * np.random.rand(100, 1)
# vecteur clonne de target
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_

```

```python
X_test = np.array([[0], [2]])
X_test_b = np.c_[np.ones((2, 1)), X_test]

y_predict = lin_reg.predict(X_test)
```

```python
plt.plot(X_test, y_predict, "r-", label="Predictions")
plt.plot(X, y, "b.", label="Donn√©es d'entrainement")
plt.xlabel("X - features")
plt.ylabel("y - target")
plt.grid(True)
plt.legend()
# fixe les limites des axes x et y
plt.axis([0, 2, 0, 15])
```
